```python
import torch
import paddle

print("--- Hardware Detection ---")

# Check for CUDA (NVIDIA GPU)
cuda_available = torch.cuda.is_available()
print(f"CUDA (NVIDIA GPU) Available: {cuda_available}")
if cuda_available:
    print(f"GPU Device Name: {torch.cuda.get_device_name(0)}")

# Check PaddlePaddle hardware binding
paddle_device = paddle.device.get_device()
print(f"PaddlePaddle Default Device: {paddle_device}")

# Summary Advice
if "gpu" in paddle_device or cuda_available:
    print("\n🚀 Performance Status: Optimal! You have GPU acceleration active.")
else:
    print("\n⚠️ Performance Status: CPU mode. Execution will work, but LLM & OCR tasks will be slower.")
```

```python
import subprocess
import time

# Spin up the Ollama background daemon
print("Starting Ollama background service...")
subprocess.Popen(["ollama", "serve"], stdout=subprocess.DEVNULL, stderr=subprocess.DEVNULL)

# Give the server 5 seconds to wake up and bind to port 11434
time.sleep(5)
print("Ollama service initialized.")
```
